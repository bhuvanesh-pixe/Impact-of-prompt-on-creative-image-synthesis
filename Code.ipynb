```python
from sentence_transformers import SentenceTransformer, util
from PIL import Image

#Load CLIP model
model = SentenceTransformer('clip-ViT-B-32')

#Encode an image:
img_emb = model.encode(Image.open('/content/farmers market.webp'))

#Encode text descriptions
text_emb = model.encode('A photorealistic image of a bustling farmers market, overflowing with fresh fruits and vegetables.')

#Compute cosine similarities
cos_scores = util.cos_sim(img_emb, text_emb)
print(cos_scores)
```

```python
from sentence_transformers import SentenceTransformer, util
from PIL import Image
import torch

# Load the models
img_model = SentenceTransformer('clip-ViT-B-32')
text_model = SentenceTransformer('sentence-transformers/clip-ViT-B-32-multilingual-v1')

def load_image(image_path):
    return Image.open(image_path)

# Define your groups of images and prompts
groups = {
    'simple': {
        'images': [
            '/content/drive/MyDrive/Simple/Simple1.webp',
            '/content/drive/MyDrive/Simple/simple2.webp',
            '/content/drive/MyDrive/Simple/Simple3.webp',
            '/content/drive/MyDrive/Simple/Simple4.webp',
            '/content/drive/MyDrive/Simple/Simple5.webp',
            '/content/drive/MyDrive/Simple/Simple6.webp',
            '/content/drive/MyDrive/Simple/Simple7.webp',
            '/content/drive/MyDrive/Simple/Simple8.webp',
            '/content/drive/MyDrive/Simple/Simple9.webp',
            '/content/drive/MyDrive/Simple/Simple10.webp'
        ],
        'texts': [
            "A busy urban coffee shop with people reading and working on laptops.",
            "A serene lake surrounded by mountains at sunrise.",
            "A colorful butterfly on a blooming flower in a garden.",
            "A snowy park with children building a snowman.",
            "A sunset view from a beach with palm trees.",
            "A modern kitchen with stainless steel appliances and a fruit bowl on the counter.",
            "A small bookstore with shelves packed with books and a cozy reading nook.",
            "A public library with rows of books and people browsing.",
            "A playground with children playing on a sunny day.",
            "A night scene of a lit-up city street with cars and pedestrians."
        ]
    },
    'historical': {
        'images': [
            '/content/drive/MyDrive/Historic/Historic1.webp',
            '/content/drive/MyDrive/Historic/Historic2.webp',
            '/content/drive/MyDrive/Historic/Historic3.webp',
            '/content/drive/MyDrive/Historic/Historic4.webp',
            '/content/drive/MyDrive/Historic/Historic5.webp',
            '/content/drive/MyDrive/Historic/Historic6.webp',
            '/content/drive/MyDrive/Historic/Historic7.webp',
            '/content/drive/MyDrive/Historic/Historic8.webp',
            '/content/drive/MyDrive/Historic/Historic9.webp',
            '/content/drive/MyDrive/Historic/Historic10.webp'
        ],
        'texts': [
            "A medieval knight tending to his horse in a stable, with a castle in the background.",
            "A bustling silk market on the Silk Road during the height of its use, with traders from different cultures.",
            "Pilgrims landing at Plymouth Rock, depicted with historical accuracy and detail.",
            "A scene from a Pharaoh's palace in Ancient Egypt, complete with hieroglyphics and servants.",
            "A festive scene in a Renaissance fair, with jesters, musicians, and townsfolk.",
            "A bustling street scene in Victorian London, with horse-drawn carriages and pedestrians in period attire.",
            "Aztec warriors preparing for a ceremonial dance in front of a step pyramid.",
            "French Impressionist painters capturing a lively Parisian café scene on canvas.",
            "A Han Dynasty scholar writing calligraphy in a tranquil garden.",
            "Viking explorers landing on the shores of Greenland, depicted with historically accurate ships and gear."
        ]
    },
    'linguistic': {
        'images': [
            '/content/drive/MyDrive/Linguistic/Ling1.webp',
            '/content/drive/MyDrive/Linguistic/Ling2.webp',
            '/content/drive/MyDrive/Linguistic/Ling3.webp',
            '/content/drive/MyDrive/Linguistic/Ling4.webp',
            '/content/drive/MyDrive/Linguistic/Ling5.webp',
            '/content/drive/MyDrive/Linguistic/Ling6.webp',
            '/content/drive/MyDrive/Linguistic/Ling7.webp',
            '/content/drive/MyDrive/Linguistic/Ling8.webp',
            '/content/drive/MyDrive/Linguistic/Ling9.webp',
            '/content/drive/MyDrive/Linguistic/Ling10.webp'
        ],
        'texts': [
            "A staircase made of books leading to the stars, symbolizing the journey of knowledge.",
            "A detailed cross-section of a computer showing all internal parts, with labels for each component.",
            "An abandoned house overgrown with vines, evoking a sense of nostalgia and loss.",
            "A city skyline at dusk, glowing like a necklace of diamonds.",
            "A forest where trees whisper secrets to each other under the moonlight.",
            "A clock with wings flying over a cityscape, symbolizing the fleeting nature of time.",
            "A botanical illustration showing the anatomy of a tropical flower, with detailed labels for each part.",
            "A cozy winter cabin at night, softly lit from within, surrounded by freshly fallen snow, evoking warmth and isolation.",
            "A desert oasis that shines like an emerald in the midst of golden sands."
            "An old library where the books feel alive, each spine containing a universe waiting to be explored."
        ]
    }
}

# Function to calculate average scores for each group
def calculate_average_scores(groups):
    for group_name, data in groups.items():
        images = [load_image(img) for img in data['images']]
        img_embeddings = img_model.encode(images, convert_to_tensor=True)
        text_embeddings = img_model.encode(data['texts'], convert_to_tensor=True)

        # Compute cosine similarities and average score
        cos_sim = util.cos_sim(text_embeddings, img_embeddings)
        average_score = torch.mean(cos_sim).item()

        print(f"Group: {group_name}")
        print(f"Average CLIP Score: {average_score}\n")

# Call the function
calculate_average_scores(groups)
```
